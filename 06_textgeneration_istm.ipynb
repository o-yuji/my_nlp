```python
from keras.models import Sequential
from keras.layers import Dense,Activation,LSTM
from keras.optimizers import RMSprop
from keras.utils.data_utils import get_file
import numpy as np
import random
import sys
```

```python
path = './data/ningen_shikkaku.txt'
```

```python
bindata = open(path,"rb").read()
text = bindata.decode("shift_jis")
print('Size of text: ',len(text))
```

```python
chars = sorted(list(set(text)))
print('Total chars: ', len(chars))
```

```python
char_indices = dict((val,key) for key,val in enumerate(chars))
```

```python
indices_char = dict((key,val) for key,val in enumerate(chars))
```

```python
maxlen = 40
step = 3
sentences = []
next_chars = []
```

```python
for i in range(0,len(text) - maxlen,step):
  sentences.append(text[i:i+maxlen])
  next_chars.append(text[i + maxlen])
```

```python
print('Number of sentences: ',len(sentences))
```

＃テキストのベクトル化

```python
X = np.zeros((len(sentences),maxlen,len(chars)),dtype=np.bool_)
```

```python
y = np.zeros((len(sentences),len(chars)),dtype=np.bool_)
```

```python
for key1,sentence in enumerate(sentences):
    for key2,char in enumerate(sentence):
        X[key1,key2,char_indices[char]] = 1
    y[key1,char_indices[next_chars[key1]]] = 1
```

#モデルを定義

```python
model = Sequential()
model.add(LSTM(128,input_shape=(maxlen,len(chars))))
model.add(Dense(len(chars)))
model.add(Activation('softmax'))
optimizer = RMSprop(learning_rate=0.01)
model.compile(loss='categorical_crossentropy',optimizer=optimizer)
```

```python
def sample(preds,temperature=1.0):
    preds = np.array(preds).astype('float64')
    preds = np.log(preds)/temperature
    exp_preds = np.exp(preds)
    preds = exp_preds/np.sum(exp_preds)
    probas = np.random.multinomial(1,preds,1)
    return np.argmax(preds)
```
